## 1. numpy, matplotlib.pyplot, pickle 라이브러리 추가

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

## 2. 바이너리 파일(train, test)를 가져와서 변수에 저장

```python
def file_to_var():
    with open('train.bin', 'rb') as f1:
        train = pickle.load(f1)
    
    with open('test.bin', 'rb') as f2:
        test = pickle.load(f2)
    return train, test
```

```python
train, test = file_to_var()
```

## 3. train이랑 test 확인

```python
print(type(train), type(test))
print(len(train), len(test))
print(type(train[0]))
print(len(train[0]))
print(type(train[0][0]))
print(train[0][0].shape)
print(train[0][0].dtype)
print(train[0][0])
plt.imshow(train[0][0],'gray')
```

## 4. train 셋의 앞에서 임의의 숫자개, test 셋의 앞에서 100개 뽑아서 사용하기!!

```python
def create_sets(train, test, num):
    trainSet = []
    testSet = []
    for i in range(10):
            trainSet.append(train[i][0:num])
            testSet.append(test[i][0:100])
    return trainSet, testSet
```

```python
trainSet, testSet = create_sets(train, test, 200)
print(type(trainSet), type(testSet))
print(len(trainSet), len(testSet))
print(type(trainSet[0]), type(testSet[0]))
print(len(trainSet[0]), len(testSet[0]))
print(type(trainSet[0][0]), type(testSet[0][0]))
print(trainSet[0][0].shape, testSet[0][99].shape)
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(trainSet[i][0], 'gray')
```

## 5. 클래스 (각 숫자당) 평균을 내준뒤 템플릿을 생성하기
## 템플릿은 리스트이고 각 리스트 값에는 한개의 매트릭스가 저장된다

```python
def create_template(trainSet):
    tmpl = np.zeros((28,200))
    for i in range(10):
        tmp = np.array(trainSet[i])
        tmpl[:,i*28:(i+1)*28] = np.mean(tmp, axis=0)
```
